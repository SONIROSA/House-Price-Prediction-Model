```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import scipy as sp
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
#Applying UniVariant Linear Regression Here
#On the Basis of No. of Bedrooms we are finding out the Price
#Bedroom will be Independent Variable & Price to be Dependent here
```

```python
y = df['bedrooms']
y =  np.array(y).reshape(-1,1)
```

```python
print(y)
```

```python
x = df['price']
```

```python
lm = LinearRegression()
o = lm.fit(y,x)
```

```python
B = o.predict(y)
B
```

```python
df['Predict_Price'] = B
df
```

```python
diff_price = df['Predict_Price']-df['price']
```

```python
df['Diff'] = diff_price
```

```python
Name = ['price','Predict_Price','Diff']
```

```python
Name = df[Name]
Name.round()
```

```python
Per_diff = Name['Diff']/Name['price']*100
```

```python
Name['Per_diff'] = Per_diff
```

```python
Name = Name.round().head()
```

```python
Name.agg({'price' : ['min','max'],'Predict_Price' : ['min','max']})
```
